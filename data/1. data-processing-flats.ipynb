```python
#https://www.thisdot.co/blog/understanding-regex
#Replace one or more(+) number (\d) followed by "." followed by numbers(\d+) followed by single space (?\s) followed by single star(?*) --> Replace with empty string ''
df['society'] = df['society'].apply(lambda x:re.sub(r'\d+(\.\d+)?\s?★','', str(x)))

df['society'] = df['society'].str.lower()

df['society'] = df['society'].apply(lambda x:str(x).strip())
```

```python
def lac_to_crore(amount):
    try:
        if type(amount) == float:
            return amount
        if str(amount[1]).lower()=='crore':
            return round(float(amount[0]), 2)
        return round(float(amount[0])/100, 2)
    except:
        print("Error value : ",amount)
        
```

```python
df[df['price'] == 'Price on Request']
```

```python
df = df[df['price'] != 'Price on Request']
df = df[df['price'] != 'price']
```

```python
df['price'] = df['price'].str.split(' ').apply(lac_to_crore)
```

```python
def update_floor(x):
    x = str(x)
    if len(x) <= 2:
        print("Error 1:",x)
        return
    try:
        #print(x[:-4])
        x = int(x[:-4].strip())
    except:
        print("Error 2:",x[:-4])

    return x
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).str.replace('Ground','0th  ').str.replace('Lower','0th  ').str.replace('Basement','-1th  ').apply(update_floor).astype(int)
```
